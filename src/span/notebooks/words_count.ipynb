```python
import json
from tqdm import tqdm
import numpy as np
import pandas as pd

# with open("multidoc2dial_dial_train.json", 'r') as f:
with open("..\\..\\..\\dataset\\multidoc2dial\\v1.0\\multidoc2dial_doc.json", 'r') as f:
    multidoc2dial_doc = json.load(f)['doc_data']

full_docs = ''

for domain in multidoc2dial_doc:
    for doc_title in tqdm(multidoc2dial_doc[domain], desc=domain):
        doc_text = multidoc2dial_doc[domain][doc_title]['doc_text']
        full_docs += doc_text + ' '
```

```python
words = full_docs.lower().split()
print(f"words count: {len(words)}")

word_counts = pd.value_counts(np.array(words))
```
